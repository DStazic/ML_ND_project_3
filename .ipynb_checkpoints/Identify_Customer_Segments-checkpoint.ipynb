```python
azdias_copy = azdias.copy()
```

```python
# Identify missing or unknown data values and convert them to NaNs.
for col in columns_dataset:  
    nan_encoded = feat_info[feat_info["attribute"]==col]["missing_or_unknown"].iloc[0]
    nan_decoded = nan_encoded.split("[")[1].split("]")[0].split(",")
    azdias_copy[col] = np.where(azdias_copy[col].isin(nan_decoded), np.nan, azdias_copy[col])
```

### Step 1.3: Create a Cleaning Function

Even though you've finished cleaning up the general population demographics data, it's important to look ahead to the future and realize that you'll need to perform the same cleaning steps on the customer demographics data. In this substep, complete the function below to execute the main feature selection, encoding, and re-engineering steps you performed above. Then, when it comes to looking at the customer data in Step 3, you can just run this function on that DataFrame to get the trimmed dataset in a single step.

```python
def clean_data(df, filter_row_nan=False):
    """
    Perform feature trimming, re-encoding, and engineering for demographics
    data
    
    INPUT: Demographics DataFrame
    OUTPUT: Trimmed and cleaned demographics DataFrame
    """
    df_copy = df.copy()
    # Put in code here to execute all main cleaning steps:
    # convert missing value codes into NaNs, ...
    for col in df_copy.columns:  
        nan_encoded = feat_info[feat_info["attribute"]==col]["missing_or_unknown"].iloc[0]
        nan_decoded = nan_encoded.split("[")[1].split("]")[0].split(",")
        df_copy[col] = np.where(df_copy[col].isin(nan_decoded), np.nan, df_copy[col])
    
    # remove selected columns and rows, ...
    # -> removal due to: 1. NaN percentage by column 2. high number of feature levels/values
    columns_drop=["TITEL_KZ", "AGER_TYP", "KK_KUNDENTYP", "GEBURTSJAHR", "KBA05_BAUMAX", "ALTER_HH",
                 "CAMEO_DEUG_2015", "LP_STATUS_FEIN", "LP_FAMILIE_FEIN", "GFK_URLAUBERTYP", "CAMEO_DEU_2015"]
    df_copy.drop(columns=columns_drop, inplace=True)
    
    if filter_row_nan:
        # select, re-encode, and engineer column values.
        # select rows below row NaN threshold
        nan_row = df_copy.isna().sum(axis=1)
        df_copy = df_copy[nan_row<=16]
    
    # create dict mappings
    orig_val = []
    dec_set = [[v]*5 for v in range(10,60,10) ]
    for dec_val_set in dec_set:
        add_v = 1
        for dec_val in dec_val_set:
            orig_val.append(dec_val+add_v)
            add_v+=1
            
    cameo_wealth_mapping = {v:np.array(dec_set).flatten()[idx] for idx,v in enumerate(orig_val)}        
    cameo_ls_mapping = {v:v%np.array(dec_set).flatten()[idx] for idx,v in enumerate(orig_val)}
    
    prae_jgd_decade_mapping = {1:40, 2:40, 3:50, 4:50, 5:60, 6:60, 7:60, 
                               8:70, 9:70, 10:80, 11:80, 12:80, 13:80, 14:90, 15:90}
    prae_jgd_movement_mapping = {1:0, 2:1, 3:0, 4:1, 5:0, 6:1, 7:1, 8:0, 
                                 9:1, 10:0, 11:1, 12:0, 13:1, 14:0, 15:1}
    
    residential_area_type_mapping = {0:1, 1:1, 2:1, 3:1, 4:1, 5:1, 7:0, 8:0}
    residential_area_quality_mapping = {7:0, 8:0}
    
    # engineer new columns
    df_copy["PRAEGENDE_JUGENDJAHRE_DECADE"] = df_copy["PRAEGENDE_JUGENDJAHRE"].dropna().apply(lambda col: prae_jgd_decade_mapping[float(col)])
    df_copy["PRAEGENDE_JUGENDJAHRE_MOV"] = df_copy["PRAEGENDE_JUGENDJAHRE"].dropna().apply(lambda col: prae_jgd_movement_mapping[float(col)])
    df_copy["CAMEO_INTL_2015_WEALTH"] = df_copy["CAMEO_INTL_2015"].dropna().apply(lambda col: cameo_wealth_mapping[float(col)])
    df_copy["CAMEO_INTL_2015_LIFE"] = df_copy["CAMEO_INTL_2015"].dropna().apply(lambda col: cameo_ls_mapping[float(col)])
    df_copy["WOHNLAGE_TYPE"] = df_copy["WOHNLAGE"].dropna().apply(lambda col: residential_area_type_mapping[float(col)])
    df_copy["WOHNLAGE_QUALITY"] = df_copy["WOHNLAGE"].dropna().apply(lambda col: residential_area_quality_mapping[float(col)]\
                                           if col in [7,8] else col)
    
    # encode categorical features as dummy variables
    columns_categorical = feat_info.loc[feat_info["type"]=="categorical", "attribute"]
    columns_reencode = set(columns_categorical).difference(set(columns_drop))
    df_copy = pd.get_dummies(df_copy, columns=columns_reencode)
    
    # Return the cleaned dataframe.
    df_copy.drop(columns=["PRAEGENDE_JUGENDJAHRE", "CAMEO_INTL_2015", "WOHNLAGE", 
                          "LP_LEBENSPHASE_FEIN", "LP_LEBENSPHASE_GROB"], inplace=True)
    df_copy.reset_index(inplace=True, drop=True)      
        
    return df_copy
    
```

### Imputation strategy
Importantly and as stated in the section above (1.1.3: Assess Missing Data in Each Row), data seems not to be missing completely at random, but conditional on a subset of feature variables. Consequently, simply removing missing values would result in biased downstream analysis.  
I decided to keep observations that were previously filtered due to specified NaN by row threshold (see (1.1.3: Assess Missing Data in Each Row) and apply the imputation strategy described in the following section.

```python
# recover the dataset
azidas_reset = clean_data(azdias)
```

I will not conduct mean/mode/median imputation, since it will likely result in strong bias, especially if applying unconditional mean imputation across all feature levels for each column (https://academic.oup.com/aje/article/160/1/34/107200, https://stats.stackexchange.com/questions/303722/methods-to-work-around-the-problem-of-missing-data-in-machine-learning). Predictive mean matching will be used because the range of imputed values will encompass the set of observed values only, thus, preserving the original category values for categorical/ordinal features. Moreover, multiple imputation steps will be conducted using the MICE algorithm, which essentially will better capture the uncertainty of imputed values. (https://www.researchgate.net/publication/44203418_MICE_Multivariate_Imputation_by_Chained_Equations_in_R, https://towardsdatascience.com/6-different-ways-to-compensate-for-missing-values-data-imputation-with-examples-6022d9ca0779, https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3074241/). Essentially, 5 imputations steps will be conducted and imputed data will either be combined using mode selection, i.e. imputed value with highest count in the imputed subsets will be selected. A tie/draw (same number of counts in multiple imputation subsets) will be resolved by randomly selecting one value from the 5 imputed subsets.  

For each feature to be imputed I will select 20 predictor variables. Using the entire feature set (114 features in azidas_reset) would simply be computationally expensive. Our goal is to capture features with highest predicting power only within the scope of the input dataset, i.e. model generalization and unbiased prediction of new and unseen data is not the objective. As such, we don't need to account for overfitting, from which a model might suffer if using too many predictor variables (curse of dimensionality). Feature selection will be conducted using the random forrest classifier, which inherently ranks features according to how well they are able to discriminate the corresponding classes in the dataset.

### rename dummy-encoded features

For some reason, dummy-encoded categorical features result in the imputation module MICE from statsmodel to throw an error. The cognate error message indicates that this is due to the categorical levels encoded in the transformed features as a float, i.e. LP_STATUS_GROB_1.0, LP_STATUS_GROB_2.0, etc.. As such let's change feature names to encode corresponding level representation as int instead, i.e. LP_STATUS_GROB_1, LP_STATUS_GROB_2.

```python
def rename_mapping(df):
    rename_mapping_summary = defaultdict()
    for col in df.columns:
        col_new = col.split(".")
        if len(col_new) > 1:
            rename_mapping_summary[col] = col_new[0]
    return rename_mapping_summary        
```

### Select feature-set for each feature to be imputed

```python
def feature_selection(y_var, df, number_feat=20):
    """
    fit default random forrest classifier and
    """
    clf = RandomForestClassifier()
    
    feature_selection_df = df.copy()
    feature_selection_df = feature_selection_df.dropna()
    y = feature_selection_df[y_var]
    X_col = [col for col in feature_selection_df.columns.difference([y_var])]
    X = feature_selection_df[X_col]
    clf.fit(X.values, y.values)
    
    ranking= sorted([feat_importance for feat_importance\
                     in zip (X_col, clf.feature_importances_)], key=lambda v: v[1], reverse=True)
    return np.array(ranking)[:,0][:number_feat]
```

```python
predictor_var_sets = defaultdict()
for feat in feature_with_nan:
    feature_set = feature_selection(feat, azidas_reset)
    predictor_var_sets[feat] = feature_set
```

### Perform imputation
Use a random sample of 200000 cases for each imputation. This is simply to speed up computation 

```python
def impute(predictor_var_sets, df, imputation_model=sm.OLS, **options):
    """
    """
    df_copy = df.copy()
    for resp_var in predictor_var_sets.keys():
        if options.get("single_imputation"):
            if resp_var != options.get("single_imputation"):
                continue
        start_time = time.time()
        # select subset
        selected_features = list(predictor_var_sets[resp_var])+[resp_var]
        df_subset = df_copy[selected_features]
        
        # group indices by missing/not missing
        nan_ind = df.index[df[resp_var].isnull()]
        print(f"Number of missing values: {len(nan_ind)}")
        observed_ind = df.index[df[resp_var].notnull()]
        if options.get("sample"):
            sample_size = options.get("sample")
            
            # combine all missing samples/rows with a subset of observed 
            observed_ind_sample = random.sample(list(observed_ind),sample_size)
            #sample_selection = df_subset.loc[list(nan_ind)+observed_ind_sample]
            df_subset = df_subset.loc[list(nan_ind)+observed_ind_sample]
        
        print(f"sample size observed: {df_subset[resp_var].notnull().sum()}\n total size: {len(df_subset)}")
        # convert data
        #imp_data = mice.MICEData(sample_selection[selected_features])
        imp_data = mice.MICEData(df_subset[selected_features])
        # define imputation model
        pred_var = " + ".join(predictor_var_sets[resp_var])
        model = resp_var + "~" + pred_var
        print(model)
        # execute imputation
        mice_model = mice.MICE(model, imputation_model, imp_data)

        results = []
        for k in range(5):
            x = mice_model.next_sample()
            imputation_results = imp_data.data.copy()
            imputation_results = imputation_results[resp_var].values
            results.append(imputation_results)
        print(f"proccessing time col {resp_var}: {(time.time()-start_time)/60}min")         
        results = np.array(results)
        final_imputation = []
        for idx in range(len(nan_ind)):
            counter = Counter(results[:,idx])
            mode = counter.most_common(1)[0][0]
            final_imputation.append(mode)
        df_copy.loc[nan_ind, resp_var] = final_imputation
        
    return df_copy
```

Categorical/ordinal features have implications on the imputation strategy. Simply methods like mean/median/mode imputation are inherently biased. Moreover they are very likely to further increase bias in imputed values, since results need to be converted (rounding) from a float representation to integer numbers that actually represent different classes for each feature. This is also true for the default MICE imputation, since predictions of dependent variables are regression-based. Overall mean/median/mode imputation method would always underestimate underrepresented class variables in case of an imbalanced class distribution. The default MICE method would better account for class imbalance, but nevertheless can corrupt imputation by "class switch", i.e true class variable is 4 but the imputed value is 4.5, which results in final class prediction to be 5.  

As a consequence, either classification via supervised machine learning or predictive mean matching in combination with MICE is a good approach to reduce bias in imputation.

```python
# Investigate the variance accounted for by each principal component.
def plot_variance(pca):
    
    pca_var = pd.DataFrame(pca.explained_variance_ratio_.reshape(1,114), 
                       columns=[f"pc_{idx}" for idx in range(len(pca.explained_variance_ratio_))])
    cumulative_var = pca_var.T.cumsum()
    cumulative_var.rename(columns={0: "cumvar"}, inplace=True)
    cumulative_var["cumvar_before"] = cumulative_var.shift().fillna(0)
    
    pc_90 = cumulative_var[cumulative_var[0]>=0.9].head(1).index.tolist()[0]
    pc_99 = cumulative_var[cumulative_var[0]>=0.99].head(1).index.tolist()[0]
    
    
    
    fig, ax = plt.subplots(1,2, figsize=(15,8))
    
    ax[0].plot(pca_var.T[0].values)
    ax[1].plot(cumulative_var[0].values)
    ax[0].set_xlabel("principle component", size=14)
    ax[1].set_xlabel("principle component", size=14)
    ax[0].set_ylabel("variance", size=14)
    ax[1].set_ylabel("variance (cumulative)", size=14)
    ax[0].set_title("variance explained by principle component", size=16)
    ax[1].set_title("cumulative variance explained", size=16)
    ax[1].hlines(0.9, 0, len(pca_var.columns), color="red", linestyle=("--"))
    ax[1].hlines(0.99, 0, len(pca_var.columns), color="red", linestyle=("--"))
    ax[1].text(0, 0.92, f"90% variance explained (pc_1-{pc_90})")
    ax[1].text(0, 1.01, f"99% variance explained (pc_1-{pc_99})")
    plt.show()

plot_variance(pca)
```

```python
# Compare the proportion of data in each cluster for the customer data to the
# proportion of data in each cluster for the general population.

def plot_cluster_distribution(cluster_set, dataset, x_label, ax=None, **options):
    if not ax:
        fig, ax = plt.subplots(1,1, figsize=(10,5))
    counts = np.bincount(cluster_set)
    rel_idx = np.where(counts>0)[0]
    counts = counts[rel_idx]
    ax2 = ax.twinx()
    
    ax.bar(range(len(counts)), (counts/(sum(counts)))*100, align='center')
    ax2.plot(range(len(counts)), counts,marker='o', c='navy', linewidth=4)
    
    ax.set(xticks=range(len(counts)), xlim=[-1, len(counts)])
    ax.set_xticklabels(rel_idx)
    
    ax.set_ylabel("Percentage of population by cluster", size=14)
    ax2.set_ylabel("Counts by cluster", size=14)
    ax.set_xlabel(f"{x_label}", size=14)
    if options.get("feature"):
        ax.set_title(f"{options.get('feature')}")
    
    plt.suptitle(f"Distribution of cluster size ({dataset})", size=16)
```

```python
plot_cluster_distribution(cluster, dataset="general population", x_label="Cluster number")
```

```python
plot_cluster_distribution(customers_cluster, dataset="customer", x_label="Cluster number")
```
